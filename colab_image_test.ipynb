```python
# Cell 1: Install deps (run once), then load model and generate one image
!pip install -q diffusers transformers accelerate torch

import torch
from diffusers import StableDiffusionPipeline

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using: {device}")

# Use a small, fast model (swap for any Hugging Face model ID to test others)
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to(device)

prompt = "Breaking news scene, professional photojournalism, dramatic lighting"
image = pipe(prompt, num_inference_steps=25).images[0]
image  # display in notebook
```

```python
# Cell 2: Save image to file (in Colab: Files panel or download from runtime)
out_path = "colab_generated_image.jpg"
image.save(out_path)
print(f"Saved: {out_path}")

# Optional: force download in Colab
try:
    from google.colab import files
    files.download(out_path)
except ImportError:
    pass
```
